```python
import pandas as pd

# Specifica il percorso al tuo file
file_path = 'dati-classifica-sanremo-1951-2023.xlsx'

# Carica solo il foglio chiamato "Python"
df_python = pd.read_excel(file_path, sheet_name='Python')
df = pd.read_excel('dati-canzoni-spotify-sanremo-1951-2023.xlsx')
```

```python
# Copie di lavoro per evitare modifiche agli originali
df_python_clean = df_python.copy()
df_clean = df.copy()

# Normalizza le colonne chiave per confronti accurati (rimuove spazi e mette in minuscolo)
for col in ['Canzone', 'Artista']:
    df_python_clean[col] = df_python_clean[col].str.strip().str.lower()
    df_clean[col] = df_clean[col].str.strip().str.lower()

# Esegui il merge mantenendo tutte le colonne di df_python + aggiunta di Popolarità
df_python_clean = df_python_clean.merge(
    df_clean[['Artista', 'Canzone', 'Popolarità' ]],
    on=['Artista', 'Canzone'],
    how='left'
)

# Visualizza le prime righe per controllo
print(df_python_clean)
```

```python
df_python_clean.to_excel('dati-classifica-popolarita-sanremo-1951-2023.xlsx',index=False)
```

```python
#Iteriamo ogni colonna del DataFrame
for column in df_python_clean.columns:
    #Somiamo i valori nulli (i dati booleani vengono interpretati come 1 e 0)
    nan_count = df_python_clean[column].isna().sum()
    #calcoliamo la percentuale
    nan_percentage = round((nan_count/df_python_clean.shape[0])*100, 2)
    print(f'{column} contains {nan_count} NaN values, {nan_percentage}% of all rows.')
```
